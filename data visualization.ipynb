```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# Paths to the image and mask directories
image_dir = r'C:\Users\HP\Desktop\My Papers\Conferences\FCN8\Data\Train\Images\aeroplane'
mask_dir = r'C:\Users\HP\Desktop\My Papers\Conferences\FCN8\Data\Train\Masks\aeroplane'

# Get a list of image and mask file names
image_files = os.listdir(image_dir)
mask_files = os.listdir(mask_dir)

# Choose 1 random index
index = np.random.choice(len(image_files))

# Initialize subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Load image and mask
image_path = os.path.join(image_dir, image_files[index])
mask_path = os.path.join(mask_dir, mask_files[index])

# Plot original image
axes[0].imshow(Image.open(image_path))
axes[0].set_title('Image')

# Plot ground truth mask
axes[1].imshow(Image.open(mask_path), cmap='gray')
axes[1].set_title('Ground Truth Mask')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
